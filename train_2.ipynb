```python
import re
def temizle(metin):
  # harfleri küçük yap
  metin = re.sub('İ', 'i', metin)
  metin = metin.lower()
  # yalnızca harf ve boşluklar (whitespace character) kalsın
  metin = re.sub('[^\w\s]', ' ', metin)
  metin = re.sub('\d', '', metin)
   # ardı sıra birden fazla boşluğu bire indir
  metin = re.sub('\s+', ' ', metin).strip()
  return metin
```

```python
import numpy as np
```

```python
def get_word2vec_embedding(text):
    words = temizle(text).split()
    embeddings = [w2v_model.wv[word] for word in words if word in w2v_model.wv]
    return np.mean(embeddings, axis=0) if embeddings else None
```

```python
pe_w2v = [get_word2vec_embedding(anlam) for anlam in anlamlar]
```

```python
from sentence_transformers import util
def get_best_indices(query_embedding, proverb_embeddings):
    similarity_scores = util.pytorch_cos_sim(query_embedding, proverb_embeddings)[0]
    return similarity_scores.topk(10).indices
```

```python
query = 'birikmek biriktirmek'

qe_w2v = get_word2vec_embedding(temizle(query))
```

```python
df_tw.iloc[get_best_indices(qe_w2v, pe_w2v)]
```

```python
for i in get_best_indices(qe_w2v, pe_w2v):
    print(int(i), atasozleri.iloc[int(i)])
    print(anlamlar.iloc[int(i)])
```

```python
from sentence_transformers import SentenceTransformer
model = SentenceTransformer('dbmdz/bert-base-turkish-uncased')
```

```python
sentences = anlamlar.apply(lambda x: temizle(x)).values
pe_bert = model.encode(sentences)
```

```python
query = 'birikmek biriktirmek'

qe_bert = model.encode(query)
```

```python
df_tw.iloc[get_best_indices(qe_bert, pe_bert)]
```
